```python
class Casa:
  """
  Construir el constructor
  argumentos:
   color(str): color de la casa
   numero_habitaciones(int): el numero de habitaciones de la casa
   numero_baos(int): el numero de baos de la casa
   estacionamento(bool): si tiene estacionamiento o no
   tiene_patio(bool): si tiene patio o no
  """
  def __init__(self,color,numero_habitaciones,numero_baos,estacionamiento,tiene_patio):
    self.color = color
    self.numero_habitaciones = numero_habitaciones
    self.numero_baos = numero_baos
    self.estacioanmiento = estacionamiento
    self.tiene_patio = tiene_patio
  def descripcion(self):
    #Describir la casa
    descripcion=f"la casa es de color {self.color}, tiene {self.numero_habitaciones}, {self.numero_baos} baos"
    if self.estacioanmiento:
      descripcion += "y tiene estacionamiento"
    if self.tiene_patio:
      descripcion += "y tiene patio"
    return descripcion

  def pintar(self,nuevo_color):
    self.color=nuevo_color
    print(f"la casa ha sio pintada de color {self.color}")
```

```python
mi_casa=Casa("blanca",5,2,True,True)
type(mi_casa)
```

```python
print(mi_casa.descripcion())
```

```python
mi_casa.tiene_patio=False
print(mi_casa.descripcion())
```

```python
mi_casa.pintar("azul")
```

```python
class Floristeria(Casa):
  """
  Constructor de la floristeria
  Argumentos:
  color(str): color de la floristeria
  numero_habitaciones(int): el numero de habitaciones de la floristeria
  numero_baos(int): el numero de baos de la floristeria
  tipos_flores(list): lista de flores
  nombre_floristeria: mombre de la floristeria
  """
  def __init__(self, color, numero_habitaciones, numero_baos, nombre_floristeria, tipos_flores):
    super().__init__(color, numero_habitaciones, numero_baos, True, True)
    self.nombre_floristeria=nombre_floristeria
    self.tipos_flores=tipos_flores
```

```python
mi_floristeria=Floristeria("azul",1,1,"Floristeria",["petunias", "hortalizas","rosas"])
mi_floristeria.descripcion()
```
