````python
# Clase Jugador
class Jugador:
    def init(self, nombre, salud=100, poder=10):
        self.nombre = nombre
        self.salud = salud
        self.poder = poder

    def atacar(self, enemigo):
        print(f"{self.nombre} ataca a {enemigo.nombre} con {self.poder} de poder.")
        enemigo.salud -= self.poder

    def estavivo(self):
        return self.salud > 0

    def str(self):
        return f"{self.nombre} (Salud: {self.salud}, Poder: {self.poder})"

# Clase Enemigo
class Enemigo:
    def init(self, nombre, salud=50, poder=5):
        self.nombre = nombre
        self.salud = salud
        self.poder = poder

    def atacar(self, jugador):
        print(f"{self.nombre} contraataca a {jugador.nombre} causando {self.poder} de da√±o.")
        jugador.salud -= self.poder

    def estavivo(self):
        return self.salud > 0

    def str(self):
        return f"{self.nombre} (Salud: {self.salud}, Poder: {self.poder})"

# Clase Juego
class Juego:
    def init(self, jugador, enemigo):
        self.jugador = jugador
        self.enemigo = enemigo

    def iniciar(self):
        print("üéÆ El juego comienza")
        while self.jugador.estavivo() and self.enemigo.estavivo():
            self.jugador.atacar(self.enemigo)
            if self.enemigo.estavivo():
                self.enemigo.atacar(self.jugador)

            print(self.jugador)
            print(self.enemigo)
            print("-" * 30)

        if self.jugador.estavivo():
            print(f"üèÜ {self.jugador.nombre} ha ganado la batalla!")
        else:
            print(f"üíÄ {self.enemigo.nombre} ha derrotado a {self.jugador.nombre}...")

# EJEMPLO DE EJECUCI√ìN

jugador = Jugador("H√©roe", salud=100, poder=15)
enemigo = Enemigo("Goblin", salud=60, poder=8)

juego = Juego(jugador, enemigo)
juego.iniciar()
````

````python
from IPython.display import YouTubeVideo, Markdown, display

# Incrustar el video en Colab
video = YouTubeVideo("_Aycjxb9Pkg", width=640, height=360)
display(video)

# Explicaci√≥n / traducci√≥n del contenido
texto_traducido = """
**Traducci√≥n y explicaci√≥n del video ‚ÄúPython Projects ‚Äì Learn to build a game (OOP) ‚Äì Minesweeper‚Äù**

El video trata sobre c√≥mo usar Python y programaci√≥n orientada a objetos para construir un juego cl√°sico: **Buscaminas (Minesweeper)**.

---

## Temas principales

- Dise√±ar clases para representar el tablero, las celdas, las minas y la l√≥gica de juego.
- C√≥mo crear m√©todos en cada clase: inicializaci√≥n (`__init__`), revelar celdas, contar minas alrededor, bandera de celda, etc.
- Separar la l√≥gica del juego de la interfaz o presentaci√≥n.
- Uso de bucles principales y llamadas recursivas para revelar √°reas vac√≠as.

---

## C√≥digo traducido / simplificado (versi√≥n en espa√±ol)

```python
import random

class Celda:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.tiene_mina = False
        self.revelada = False
        self.banderita = False
        self.minas_vecinas = 0

    def __repr__(self):
        if self.banderita:
            return "F"
        if not self.revelada:
            return "X"
        if self.tiene_mina:
            return "*"
        return str(self.minas_vecinas)

class Tablero:
    def __init__(self, filas, columnas, num_minas):
        self.filas = filas
        self.columnas = columnas
        self.num_minas = num_minas
        self.celdas = [[Celda(r, c) for c in range(columnas)] for r in range(filas)]
        self._colocar_minas()
        self._calcular_vecinas()

    def _colocar_minas(self):
        colocadas = 0
        while colocadas < self.num_minas:
            r = random.randrange(self.filas)
            c = random.randrange(self.columnas)
            celda = self.celdas[r][c]
            if not celda.tiene_mina:
                celda.tiene_mina = True
                colocadas += 1

    def _calcular_vecinas(self):
        for r in range(self.filas):
            for c in range(self.columnas):
                if self.celdas[r][c].tiene_mina:
                    continue
                conteo = 0
                for dr in [-1, 0, 1]:
                    for dc in [-1, 0, 1]:
                        rr = r + dr
                        cc = c + dc
                        if 0 <= rr < self.filas and 0 <= cc < self.columnas:
                            if self.celdas[rr][cc].tiene_mina:
                                conteo += 1
                self.celdas[r][c].minas_vecinas = conteo

    def revelar(self, r, c):
        celda = self.celdas[r][c]
        if celda.revelada or celda.banderita:
            return
        celda.revelada = True
        if celda.minas_vecinas == 0 and not celda.tiene_mina:
            for dr in [-1, 0, 1]:
                for dc in [-1, 0, 1]:
                    rr = r + dr
                    cc = c + dc
                    if 0 <= rr < self.filas and 0 <= cc < self.columnas:
                        if not self.celdas[rr][cc].revelada:
                            self.revelar(rr, cc)

    def mostrar(self):
        for fila in self.celdas:
            print(" ".join(str(celda) for celda in fila))

# Uso b√°sico
if __name__ == "__main__":
    juego = Tablero(5, 5, 5)
    juego.mostrar()
    # luego puedes llamar a juego.revelar(r, c) y volver a mostrar()
````
