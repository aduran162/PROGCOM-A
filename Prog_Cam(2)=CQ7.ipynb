```python
# REPOSITORIO DE PALABRAS


# 1. False
# Definición: Valor booleano falso.
x = False
if not x:
    print("Ejemplo 1: x es falso")

# 2. True
# Definición: Valor booleano verdadero.
x = True
if x:
    print("Ejemplo 2: x es verdadero")

# 3. None
# Definición: Representa la ausencia de valor.
x = None
print("Ejemplo 3:", x)

# 4. and
# Definición: Operador lógico "y".
print("Ejemplo 4:", True and False)

# 5. or
# Definición: Operador lógico "o".
print("Ejemplo 5:", True or False)

# 6. not
# Definición: Operador lógico de negación.
print("Ejemplo 6:", not True)

# 7. as
# Definición: Alias al importar módulos.
import math as m
print("Ejemplo 7:", m.sqrt(16))

# 8. assert
# Definición: Verifica condiciones.
x = 10
assert x > 0
print("Ejemplo 8: assert correcto")

# 9. break
# Definición: Sale de un bucle.
for i in range(5):
    if i == 2:
        break
print("Ejemplo 9: break detenido en 2")

# 10. class
# Definición: Define una clase.
class Persona:
    pass
print("Ejemplo 10: clase creada")

# 11. continue
# Definición: Salta a la siguiente iteración.
for i in range(5):
    if i == 2:
        continue
print("Ejemplo 11: continue saltó el 2")

# 12. def
# Definición: Define una función.
def saludar():
    return "Hola"
print("Ejemplo 12:", saludar())

# 13. del
# Definición: Elimina variables o elementos.
lista = [1,2,3]
del lista[0]
print("Ejemplo 13:", lista)

# 14. elif
# Definición: Condicional extra.
x = 5
if x < 0:
    print("negativo")
elif x == 5:
    print("Ejemplo 14: cinco")

# 15. else
# Definición: Alternativa condicional.
if False:
    print("no")
else:
    print("Ejemplo 15: else usado")

# 16. except
# Definición: Captura excepciones.
try:
    5/0
except ZeroDivisionError:
    print("Ejemplo 16: excepción capturada")

# 17. finally
# Definición: Bloque que siempre se ejecuta.
try:
    x = 1/0
except:
    print("error")
finally:
    print("Ejemplo 17: finally ejecutado")

# 18. for
# Definición: Bucle iterativo.
for i in range(3):
    print("Ejemplo 18:", i)

# 19. from
# Definición: Importar algo específico.
from math import pi
print("Ejemplo 19:", pi)

# 20. global
# Definición: Declara variables globales.
def f():
    global var
    var = 10
f()
print("Ejemplo 20:", var)

# 21. if
# Definición: Condicional.
if True:
    print("Ejemplo 21: if verdadero")

# 22. import
# Definición: Importa librerías.
import random
print("Ejemplo 22:", random.randint(1,5))

# 23. in
# Definición: Verifica pertenencia.
print("Ejemplo 23:", 3 in [1,2,3])

# 24. is
# Definición: Compara identidades.
a = None
print("Ejemplo 24:", a is None)

# 25. lambda
# Definición: Crea funciones anónimas.
doble = lambda x: x*2
print("Ejemplo 25:", doble(5))

# 26. nonlocal
# Definición: Usa variables de un scope externo.
def f():
    x = "local"
    def g():
        nonlocal x
        x = "cambiado"
    g()
    return x
print("Ejemplo 26:", f())

# 27. pass
# Definición: Bloque vacío.
def vacio():
    pass
print("Ejemplo 27: pass ejecutado")

# 28. raise
# Definición: Lanza excepciones.
try:
    raise ValueError("error")
except:
    print("Ejemplo 28: excepción lanzada")

# 29. return
# Definición: Devuelve un valor en función.
def sumar(a,b):
    return a+b
print("Ejemplo 29:", sumar(2,3))

# 30. try
# Definición: Manejo de errores.
try:
    1/0
except:
    print("Ejemplo 30: error capturado")

# 31. while
# Definición: Bucle mientras se cumpla condición.
i = 0
while i < 3:
    print("Ejemplo 31:", i)
    i+=1

# 32. with
# Definición: Manejo de recursos.
with open("temp.txt","w") as f:
    f.write("hola")
print("Ejemplo 32: archivo escrito")

# 33. yield
# Definición: Genera valores en un generador.
def gen():
    yield 1
    yield 2
print("Ejemplo 33:", list(gen()))

# 34. abs
# Devuelve valor absoluto.
print("Ejemplo 34:", abs(-5))

# 35. all
print("Ejemplo 35:", all([True, True, False]))

# 36. any
print("Ejemplo 36:", any([False, False, True]))

# 37. bin
print("Ejemplo 37:", bin(10))

# 38. bool
print("Ejemplo 38:", bool(0))

# 39. chr
print("Ejemplo 39:", chr(65))

# 40. dict
d = dict(a=1,b=2)
print("Ejemplo 40:", d)

# 41. float
print("Ejemplo 41:", float(5))

# 42. int
print("Ejemplo 42:", int("10"))

# 43. len
print("Ejemplo 43:", len([1,2,3]))

# 44. list
print("Ejemplo 44:", list("abc"))

# 45. max
print("Ejemplo 45:", max([1,9,3]))

# 46. min
print("Ejemplo 46:", min([1,9,3]))

# 47. open
with open("archivo.txt","w") as f:
    f.write("test")
print("Ejemplo 47: archivo creado")

# 48. pow
print("Ejemplo 48:", pow(2,3))

# 49. print
print("Ejemplo 49: hola mundo")

# 50. range
print("Ejemplo 50:", list(range(3)))

# 51. round
print("Ejemplo 51:", round(3.1416,2))

# 52. set
print("Ejemplo 52:", set([1,2,2,3]))

# 53. sorted
print("Ejemplo 53:", sorted([3,1,2]))

# 54. str
print("Ejemplo 54:", str(123))

# 55. sum
print("Ejemplo 55:", sum([1,2,3]))

# 56. tuple
print("Ejemplo 56:", tuple([1,2,3]))

# 57. type
print("Ejemplo 57:", type(3))

# 58. zip
print("Ejemplo 58:", list(zip([1,2],[3,4])))

# 59. enumerate
print("Ejemplo 59:", list(enumerate(["a","b"])))

# 60. map
print("Ejemplo 60:", list(map(lambda x:x*2,[1,2,3])))
```
